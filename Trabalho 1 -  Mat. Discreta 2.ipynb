### Criando classe para a lista encadeada

```python
class Node:
    def __init__(self, data, nextNode = None):
        self.data = data
        self.next = nextNode
        
class LinkedList:
    def __init__(self):
        self.head = None
        self.tail = None
        
#     Encontra valores dentro da lista encadeada e retorna a sua posi√ß√£o
    def findData(self, data):
        aux = self.head
        position = 0
        
        while(aux != None):
            if aux.data == data:
                return position
            
            aux = aux.next
            position += 1
        
        return None
    
#     Adiciona novos n√≥s ao final da lista    
    def appendToTail(self, data):
        if not self.head:
            self.head = self.tail = Node(data)
        else:
            self.tail.next = Node(data)
            self.tail = self.tail.next
            
#     Remove n√≥s do final da lista
    def removeFromTail(self):
        aux = self.tail
        removedData = aux.data
        
        if not self.tail:
            print(f'A lista est√° vazia!')
        elif self.head == self.tail:
            self.head = self.tail = None
        else:
            aux2 = self.head
            
            while(aux2.next != aux):
                aux2 = aux2.next
            
            aux2.next = None
            
        del aux
    
#     Printa todos os n√≥s e suas posi√ß√µes
    def printNodes(self):
        aux = self.head
        position = 0
        
        while(aux != None):
            print(f'N√≥ {position}: {aux.data}')
            
            aux = aux.next
            position += 1
```

```python
lista = LinkedList()

lista.appendToTail(8)
lista.appendToTail(8)
lista.appendToTail(8)
lista.appendToTail(8)
lista.appendToTail(9)

lista.printNodes()
```

### Criando a tabela Hash com encadeamento exterior

```python
class HashTableExterior:
    def __init__(self, m):
        self.length = m
        self.table = [LinkedList() for i in range(0, self.length)]
        
#     Adiciona dados na tabela
    def append(self, data):
        lastThreeDigits = data % 1000
        
        index = lastThreeDigits % self.length
        
        self.table[index].appendToTail(data)
        
#     Encontra dados na tabela, retornando seu endere√ßo base e qual sua posi√ß√£o dentro do endere√ßo base
    def findData(self, data):
        lastThreeDigits = data % 1000
        
        index = lastThreeDigits % self.length
        
        position = self.table[index].findData(data)
        
        if(position):
            return print(f'{data} encontrada no endere√ßo base {index+1} na posi√ß√£o {position} da lista encadeada')
        
        return print(f'{data} n√£o encontrada')
      
#     Printa todos os dados da tabela
    def printTable(self):
        for i, eachAdress in enumerate(self.table):
            print(f'\nDados no endere√ßo base {i+1}:')
            eachAdress.printNodes()
```

### Alimentando a tabela Hash exterior com o conjunto de contas

```python
hashTableEx = HashTableExterior(9)

for conta in numeros_conta:
    hashTableEx.append(conta)
    
hashTableEx.printTable()
```

### Fazendo uma tabela Hash interior

```python
class HashNode:
    def __init__(self, data = None, position = None, nextNodeIndex = None):
        self.data = data
        self.next = nextNodeIndex
        self.position = position

class HashTableInterior:
    def __init__(self, p):
        self.baseAdressLength = p
#         Cria as p posi√ß√µes iniciais da tabela
        self.table = [HashNode() for i in range(0, self.baseAdressLength)]
        self.colisionAreaLength = 0
        
#     Adiciona dados na tabela
    def append(self, data):
        lastThreeDigits = data % 1000
        
        index = lastThreeDigits % self.baseAdressLength
        
        if not self.table[index].data:
            self.table[index].data = data
            self.table[index].position = index+1
        else:
            aux = self.table[index]
            
            while(aux.next != None):
                aux = self.table[aux.next]
                
            self.colisionAreaLength += 1
            self.table.append(HashNode(data, self.baseAdressLength + self.colisionAreaLength))
            aux.next = (self.baseAdressLength + self.colisionAreaLength) - 1
            
        
#     Encontra dados na tabela, retornando seu endere√ßo base e qual sua posi√ß√£o dentro do endere√ßo base
    def findData(self, data):
        lastThreeDigits = data % 1000
        
        index = lastThreeDigits % self.baseAdressLength
        
        
        aux = self.table[index]
            
        while(True):
            if aux.data == data:
                return print(f'{data} encontrada na posi√ß√£o {aux.position} da tabela Hash')
            
            aux = self.table[aux.next]
            
            if aux.next == None:
                break
        
        return print(f'{data} n√£o encontrada')
      
#     Printa todos os dados da tabela
    def printTable(self):
        for i, eachAdress in enumerate(self.table[:self.baseAdressLength]):
            print(f'\nDados no endere√ßo base {i+1}:')
            while(eachAdress != None):
                nextPosition = None if not eachAdress.next else eachAdress.next+1
                print(f'Valor {eachAdress.data} - Posi√ß√£o: {eachAdress.position} - Pr√≥xima posi√ß√£o: {nextPosition}')
                
                if eachAdress.next == None:
                    eachAdress = None
                else:
                    eachAdress = self.table[eachAdress.next]
```

### Alimentando a tabela Hash interior com o conjunto de contas

```python
hashTableIn = HashTableInterior(9)

for conta in numeros_conta:
    hashTableIn.append(conta)

hashTableIn.printTable()
```
